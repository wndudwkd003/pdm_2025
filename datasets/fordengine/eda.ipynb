```python
from scipy.io import arff
import numpy as np
import pandas as pd

def load_arff(path: str):
    raw_data, meta = arff.loadarff(path)
    df = pd.DataFrame(raw_data)

    # 마지막 컬럼이 라벨 (Class)
    y = df.iloc[:, -1].astype(int).values
    X = df.iloc[:, :-1].astype(float).values

    return X, y



train_path = "/ws/pdm_2025/datasets/fordengine/dataset/FordA_TRAIN.arff"
test_path  = "/ws/pdm_2025/datasets/fordengine/dataset/FordA_TEST.arff"

X_train, y_train = load_arff(train_path)
X_test,  y_test  = load_arff(test_path)

print("X_train:", X_train.shape)
print("y_train:", y_train.shape)
print("X_test:", X_test.shape)
print("y_test:", y_test.shape)
```
